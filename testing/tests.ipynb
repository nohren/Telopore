```python
import re
import matplotlib.pyplot as plt

#chromosome consistency

chr = []
seqLen = []
with open('../nanosim/CN1_telopore_sim_2_1000_aligned_reads.fasta', "r") as f:
        for line in f:
            line.strip()
            if line.startswith(">"):
                c = line.split("_")[0][1:]
                c_i = re.sub(r'[^0-9]', '', c)
                if c_i:
                    chr.append(int(c_i))
            else:
                seqLen.append(len(line))

freq = {}
for i in range(1,23):
    freq[i] = 0
for c in chr:
    if c in freq:
        freq[c] += 1
    else:
        freq[c] = 1

data = sorted([(x,f) for x,f in freq.items()])
x = [x for x,_ in data]
y = [y for _,y in data]


print(sum(seqLen)/len(seqLen))

plt.figure()

# Plot the data

plt.bar(x, y)
# Label axes
plt.xlabel("X-axis label")
plt.ylabel("Y-axis label")

# Optional title
plt.title("Chromosome representation")

# Show the plot
plt.show()



```
